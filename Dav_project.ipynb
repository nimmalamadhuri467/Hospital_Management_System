```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/data.csv")

#prompt: Display the first 10 rows of the dataset to understand its structure
df.head(10)

#prompt: Get dataset info including column names, data types, and missing values
df.info()

#prompt: Get summary statistics for numerical columns
df.describe()
```

```python
# Fill numeric columns with mean
df.fillna(df.select_dtypes(include=['number']).mean(), inplace=True)

# Fill categorical columns with mode
df.fillna(df.select_dtypes(include=['object']).mode().iloc[0], inplace=True)

# Display updated DataFrame
print(df.head(10))  # Show first 10 rows

```
